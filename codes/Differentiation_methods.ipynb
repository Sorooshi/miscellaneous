```python


import numpy as np 
import matplotlib.pyplot as plt 
from scipy.misc import derivative




np.set_printoptions(precision=5, suppress=True)


```

```python

x = np.linspace(0,5*np.pi,100)

y = np.sin(x).reshape(-1, 1)


h=1e-5

```

```python
dy_dx_true = np.cos(x)


dy_dx_forward = (np.sin(x+h) - np.sin(x))/h


dy_dx_np_grad = np.gradient(np.sin(x), axis=0)
dy_dx_sp_derive = derivative(np.sin, x, )




```

```python


plt.plot(dy_dx_true, "g-+", label="True")
plt.plot(dy_dx_forward, "b-.", label="Forward")
plt.plot(dy_dx_sp_derive, "r-*", label="Scipy Gradient")
plt.plot(dy_dx_np_grad, "m-.", label="Numpy Gradient")


plt.legend()


plt.show()

```
