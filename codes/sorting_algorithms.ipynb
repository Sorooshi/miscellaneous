# Sorting algorithms

## Select sort

```python
def select_sort(arr):
    
    arr_s = []  # sorted array
    _min = np.inf
    n = len(arr)
    
    while len(arr_s) < n:
        _min = min(arr)
        arr_s.append(_min)
        arr.remove(_min)
    return arr_s
        
        
```

```python
array = [64, 34, 25, 12, 64, 22, 11, 90]



sel_srot = select_sort(arr=array)

print("select sort result:", sel_srot)


```

## Bubble sort

```python

def bubble_sort(arr):
    n = len(arr)
    sorting = True
    
    while sorting:
        sorting = False
        print("begining:", sorting)
        for i in range(n-1):            
            if arr[i+1] < arr[i-1]:
                tmp = arr[i+1]
                arr[i+1] = arr[i]
                arr[i] = tmp 
                sorting = True
                print("if:", sorting)
        print("after for:", sorting, arr)
    return arr 
```

```python
array = [64, 34, 25, 12, 64, 22, 11, 90]



buble_sort = bubble_sort(arr=array)


print("Bubble sort result:", buble_sort)

```
