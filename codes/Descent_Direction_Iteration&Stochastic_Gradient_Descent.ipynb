```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def objective(x):
    return x ** 2
```

```python
def derivative(x):
    return x * 2
```

```python
# define range for input
r_min, r_max = -1.0, 1.0
# sample input range uniformly at 0.1 increments
inputs = np.arange(r_min, r_max+0.1, 0.1)

# compute target values
results = objective(inputs)
```

```python
# create a line plot of input vs result
plt.plot(inputs, results)
# show the plot
plt.show()
```

```python
# define range for input
bounds = np.asarray([[-1.0, 1.0]])
# define the total iterations
n_iter = 30
# define the maximum step size
step_size = 0.1

```

```python
def gradient_descent(objective, gradient, bounds, n_iter, step_size):
    # initial point
    solution = bounds[:, 0] + np.random.rand(len(bounds)) * (bounds[:, -1] - bounds[:, 0])
    for i in range(n_iter):
        # evaluate the gradient
        eval_grad = gradient(solution)
        # take a step
        solution = solution - step_size * eval_grad
        # evaluate candidate point
        eval_solution = objective(solution)
        # report progress
        print('>%d f(%s) = %.5f' % (i, solution, eval_solution))
    return solution, eval_solution
```

```python
best, score = gradient_descent(objective=objective, 
                               gradient=derivative,
                               bounds=bounds,
                               n_iter=30, 
                              step_size=1)
print('Done!')
print('f(%s) = %f' % (best, score))
```

```python
best, score = gradient_descent(objective=objective, 
                               gradient=derivative,
                               bounds=bounds,
                               n_iter=30, 
                              step_size=1e-1)
print('Done!')
print('f(%s) = %f' % (best, score))
```
